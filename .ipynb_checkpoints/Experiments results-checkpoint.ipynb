```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
python_y = [3755/60, 5187/60]
matlab_y = [2772/60, 11468/60]
N = 2

ind = np.arange(N)
width = 0.35

plt.barh(ind, python_y, width, label='Python')
plt.barh(ind + width, matlab_y, width, label='MATLAB')

plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='x', alpha=0.2)

plt.ylabel('Nodes')
plt.title('Fitness Function Run Times')

plt.yticks(ind + width / 2, ('20', '59'))
plt.xlabel('Time (minutes)')
plt.legend(loc='best')
plt.show()
```

```python
y_found_SI = [0.6721, 0.4172, 0.3121, 0.1774, 0.0605, 0.013, 0.0058, 0.0025, 0.001, 0.0002]
ground_truth = [0.6417, 0.3573, 0.2575, 0.1405, 0.0472, 0.0107, 0.0049, 0.0015, 0.00092267, 0.0002876]

width = 0.35
X = np.arange(10)

plt.bar(X, y_found_SI, width, label='Optimiser Solution')
plt.bar(X + width, ground_truth, width, label='Ground Truth')

plt.ylabel('1 - Set Ictogenicity')
plt.title('Optimiser vs Ground Truth')

plt.yticks(np.arange(0,1.1,0.1))
plt.xticks(range(1,11))
plt.xlabel('Sorted sets of nodes')
plt.legend(loc='best')
plt.show()
```
